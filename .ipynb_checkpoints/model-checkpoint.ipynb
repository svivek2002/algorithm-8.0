## Libraries import

```python
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

## Search for Target protein 

### Target search for coronavirus

```python
tg = new_client.target
tg_query = tg.search('coronavirus')
targets_df = pd.DataFrame.from_dict(tg_query)
targets_df
```

### Choose and get SARS coronavirus 3C-like proteinase bioactivity data (seventh entry).

```python
tg_selected = targets_df.target_chembl_id[6]
tg_selected
```

##### Only the coronavirus 3C-like proteinase (CHEMBL3927) bioactivity data that are given as IC 50 values in nM (nanomolar) units will be retrieved in this case.

```python
activity = new_client.activity
result = activity.filter(target_chembl_id=tg_selected).filter(standard_type="IC50")
```

```python
df = pd.DataFrame.from_dict(result)
df.head()
```

#### Save the bioactivity data to a csv file

```python
df.to_csv('data_raw.csv', index=False)
```

### Handling missing data

```python
df2 = df[df.standard_value.notna()]
df2
```

## Data Preprocessing

### 3 classes classification of compounds - Active, Inactive and Intermediate

The IC50 unit contains the bioactivity data. Compounds will be classified as **active** if their value is less than 1000 nM, and as **inactive** if it is more than 10,000 nM. We will refer to values that fall between 1,000 and 10,000 nM as  **intermediate**.

```python
compound_class = []
for i in df2.standard_value:
  if float(i) >= 10000:
    compound_class.append("inactive")
  elif float(i) <= 1000:
    compound_class.append("active")
  else:
    compound_class.append("intermediate")
```

### Extracting and combining useful columns 

```python
selection = ['molecule_chembl_id','canonical_smiles','standard_value']
df3 = df2[selection]
df3
```

```python
compound_class = pd.Series(compound_class, name='compound_class')
df4 = pd.concat([df3, compound_class], axis=1)
df4
```

### Save dataframe to CSV file

```python
df4.to_csv('data_preprocessed.csv', index=False)
```
